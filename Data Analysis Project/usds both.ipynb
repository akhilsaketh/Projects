### There is a myth that it is will be bad for people who born on 13th friday.we have to find out whether it is still following to not give births on 13th friday or not?
Data set is given in the folder of Data analyis.

# <span style="color:blue">SOLUTION:</span>

```python
#loading the file
file=open('US_births_2000-2014_SSA.csv')
```

```python
#passing the values of file to a list 'lines' by splitting
lines=[line.strip().split(',') for line in file]
```

```python
lines
```

```python
from functools import reduce
```

With the help of __*LAMBDA,REDUCE & FILTER*__ Functions,Finding out the Sum of all members born in all years as per date and as per day

```python
perweek=[]
for j in range(1,8):
    
    perweek.append(reduce (lambda x,y:int(x)+int(y),[list(filter(lambda x:x[2]=='13'and x[3]==str(j),lines))[i][4] for i in range(len(list(filter(lambda x:x[2]=='13'and x[3]==str(j),lines))))]))

print(perweek)
```

### Sum of Members born on 13th on different Weeks

### It is observed from the above results that on friday the count is not the least one among all.

###  Sum of Members born on FRIDAY on different Days

```python
perday=[]
for k in range(1,32):
    perday.append(reduce (lambda x,y:int(x)+int(y),[list(filter(lambda x:x[2]==str(k)and x[3]=='5',lines))[i][4] for i in range(len(list(filter(lambda x:x[2]==str(k)and x[3]=='5',lines))))]))
```

```python
'''In Friday of different days'''
for l in range(1,10):
    print(l,"  -",perday[l-1])
for l in range(10,32):
    print(l," -",perday[l-1])
```

### Here once again we can find out that the count on 13th is not the least one among all.
